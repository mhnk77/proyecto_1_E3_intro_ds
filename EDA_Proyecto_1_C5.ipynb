```python
import pandas as pd
```

```python
data = pd.read_csv('data/incidentes-viales-c5.csv')
data.head()
```

```python
data.dtypes
```

```python
data.nunique(axis = 0)
```

```python
data['fecha_creacion'] = pd.to_datetime(data['fecha_creacion'], format = '%d/%m/%Y', errors = 'coerce')
data['fecha_cierre'] = pd.to_datetime(data['fecha_cierre'], format = '%d/%m/%Y', errors = 'coerce')
data['delegacion_inicio'] = data.delegacion_inicio.str.lower()
data['delegacion_cierre'] = data.delegacion_cierre.str.lower()
data.drop(columns = ['geopoint'], inplace = True)
data.head()
```
